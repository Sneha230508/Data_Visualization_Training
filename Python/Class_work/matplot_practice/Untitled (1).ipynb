```python
import matplotlib.pyplot as plt

days = list(range(1,11))
students = [40,42,43,45,46,48,47,49,50,52]

plt.figure(figsize=(8,5))
plt.plot(days, students, marker='o', color='cyan', linestyle='-')
plt.title('Student Attendance Over 10 Days')
plt.xlabel('Day')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct']
rainfall = [10,15,20,30,40,80,120,100,60,30]

plt.figure(figsize=(8,5))
plt.plot(months, rainfall, marker='s', color='blue', linestyle='--')
plt.title('Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()
```
