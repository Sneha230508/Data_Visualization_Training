```python
# Import matplotlib library
import matplotlib.pyplot as plt

# List of student exam scores
scores = [45,50,55,60,62,65,68,70,72,75,78,80,83,85,90]

# Create histogram
# bins divide data into groups
plt.hist(scores, bins=5)

# Title
plt.title("Student Exam Score Distribution")

# Axis labels
plt.xlabel("Score Range")
plt.ylabel("Number of Students")

plt.show()
```

```python
# Import matplotlib.pyplot module
# This module is used to create graphs and charts in Python
import matplotlib.pyplot as plt

# Create a list of employee salaries
# These values represent salary amounts for different employees
salary = [20000, 25000, 30000, 32000, 35000, 38000, 40000, 42000, 45000, 48000]

# Create a histogram to show how salary values are distributed
# bins=5 means the data will be divided into 5 groups
plt.hist(salary, bins=5)

# Add title to describe the chart
plt.title("Employee Salary Distribution")

# Label for X-axis showing salary ranges
plt.xlabel("Salary Range")

# Label for Y-axis showing number of employees
plt.ylabel("Number of Employees")

# Display the histogram
plt.show()
```
