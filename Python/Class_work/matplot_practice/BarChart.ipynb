```python
# Import matplotlib.pyplot module and give it the short name 'plt'
# This library is used for creating graphs and charts in Python
import matplotlib.pyplot as plt

# Create a list of course names
# These values will appear on the X-axis of the bar chart
courses = ["Python", "Java", "C++", "Web", "AI"]

# Create a list showing number of students in each course
# These values will appear on the Y-axis
students = [40, 35, 25, 30, 20]

# Create the bar chart
# First argument = X-axis categories
# Second argument = Y-axis numerical values
plt.bar(courses, students)

# Add title to the graph to describe what the graph represents
plt.title("Students in Different Courses")

# Label for the X-axis
plt.xlabel("Courses")

# Label for the Y-axis
plt.ylabel("Number of Students")

# Display the graph window
plt.show()
```

```python
# Import matplotlib library
import matplotlib.pyplot as plt

# List of products (categories for X-axis)
products = ["Laptop", "Mobile", "Tablet", "Keyboard", "Mouse"]

# Sales numbers for each product (values for Y-axis)
sales = [120, 200, 90, 150, 180]

# Create bar chart to compare product sales
plt.bar(products, sales)

# Title describing the graph
plt.title("Sales by Product")

# Label for X-axis
plt.xlabel("Products")

# Label for Y-axis
plt.ylabel("Sales")

# Show the chart
plt.show()
```
