```python
# Import matplotlib.pyplot for creating charts
import matplotlib.pyplot as plt

# List of mobile brands (labels for each slice of the pie chart)
brands = ["Samsung", "Apple", "Xiaomi", "Oppo", "Others"]

# Market share percentage for each brand
share = [30, 25, 20, 15, 10]

# Create pie chart
# labels = names displayed on each slice
# autopct = display percentage values on the chart
plt.pie(share, labels=brands, autopct='%1.1f%%')

# Title of the chart
plt.title("Mobile Market Share")

# Display the chart
plt.show()
```
