```python
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales = [120,150,170,160,180,200,220,210,190,230,250,270]

plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# Multiple Line Chart – Three Products
months = ['Jan','Feb','Mar','Apr','May','Jun']

A = [10,20,30,40,50,60]
B = [15,25,35,45,55,65]
C = [20,30,40,50,60,70]

plt.plot(months, A, label="Product A")
plt.plot(months, B, label="Product B")
plt.plot(months, C, label="Product C")

plt.legend()
plt.title("Sales of Products")
plt.show()
```

```python
#Bar Chart – Students in Courses
courses = ['Python','Java','Data Science','Web Dev']
students = [40,30,25,35]

plt.bar(courses, students)
plt.title("Students Enrolled in Courses")
plt.show()
```

```python
# Horizontal Bar Chart – Population
cities = ['Delhi','Mumbai','Chennai','Kolkata','Bangalore']
population = [32,20,11,14,13]

plt.barh(cities, population)
plt.title("Population of Cities")
plt.show()
```

```python
# Stacked Bar Chart – Male & Female Students
dept = ['CSE','ECE','ME','CE']
male = [50,40,30,20]
female = [45,35,25,15]

plt.bar(dept, male, label="Male")
plt.bar(dept, female, bottom=male, label="Female")

plt.legend()
plt.title("Students in Departments")
plt.show()
```

```python
# Pie Chart – Market Share
brands = ['Apple','Samsung','Xiaomi','Oppo','Vivo']
share = [30,25,20,15,10]

plt.pie(share, labels=brands, autopct='%1.1f%%')
plt.title("Smartphone Market Share")
plt.show()
```

```python
# Exploded Pie Chart – Daily Activities
activities = ['Study','Sleep','Exercise','Entertainment']
time = [6,8,2,4]
explode = (0.1,0,0,0)

plt.pie(time, labels=activities, explode=explode, autopct='%1.1f%%')
plt.title("Daily Time Spent")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

marks = np.random.randint(40,100,100)

plt.hist(marks, bins=10)
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [2019,2020,2021,2022,2023,2024]
profit = [10,15,20,25,30,35]

plt.fill_between(years, profit)

plt.title("Yearly Profit Growth")
plt.xlabel("Years")
plt.ylabel("Profit")
plt.show()
```

```python
import matplotlib.pyplot as plt

time = ["6AM","9AM","12PM","3PM","6PM","9PM"]
consumption = [2,4,6,5,7,3]

plt.step(time, consumption)

plt.title("Electricity Consumption in a Day")
plt.xlabel("Time")
plt.ylabel("Units")
plt.show()
```

```python
import matplotlib.pyplot as plt

dept1 = [40,45,50,55,60]
dept2 = [35,40,45,50,55]
dept3 = [50,55,60,65,70]

data = [dept1, dept2, dept3]

plt.boxplot(data)
plt.title("Salary Distribution")
plt.xticks([1,2,3],["Dept1","Dept2","Dept3"])
plt.show()
```

```python
import matplotlib.pyplot as plt

classA = [60,65,70,75,80]
classB = [55,60,65,70,75]
classC = [70,75,80,85,90]

data = [classA,classB,classC]

plt.violinplot(data)
plt.title("Distribution of Exam Scores")
plt.show()
```
