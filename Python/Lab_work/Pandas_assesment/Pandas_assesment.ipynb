```
import pandas as pd

salary = [25000, 32000, 28000, 40000, 50000]
emp = ['A', 'B', 'C', 'D', 'E']

s = pd.Series(salary, index=emp)

# Increase salary by 10% where salary < 30000
s[s < 30000] = s[s < 30000] * 1.10

# Decrease salary by 5% where salary > 45000
s[s > 45000] = s[s > 45000] * 0.95

print("Updated Salary:\n", s)

# Total salary expense
print("\nTotal Salary:", s.sum())
```

```
import pandas as pd

data = [10, None, 30, None, 50, 60, None]
s = pd.Series(data)

# Count missing before
before = s.isnull().sum()

# Fill missing with mean
mean_val = s.mean()
s_filled = s.fillna(mean_val)

# Count after
after = s_filled.isnull().sum()

print("Before:", before)
print("After:", after)

# Drop remaining nulls (if any)
s_clean = s_filled.dropna()
print("\nClean Data:\n", s_clean)
```

```
import pandas as pd

marks = [95, 82, 67, 58, 73, 49, 88]
students = ['S1','S2','S3','S4','S5','S6','S7']

s = pd.Series(marks, index=students)

# Function to assign grade
def grade(m):
    if m >= 90:
        return 'A'
    elif m >= 75:
        return 'B'
    elif m >= 60:
        return 'C'
    else:
        return 'D'

# Apply grading
grades = s.apply(grade)
print("Grades:\n", grades)

# Count each grade
print("\nGrade Count:\n", grades.value_counts())
```

```
import pandas as pd

s1 = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
s2 = pd.Series([5, 15, 25], index=['B', 'C', 'D'])

# Add both Series
result = s1 + s2
print("With NaN:\n", result)

# Replace NaN with 0 and recompute
result_fixed = s1.add(s2, fill_value=0)
print("\nAfter replacing NaN:\n", result_fixed)
```

```
import pandas as pd

scores = [88, 92, 79, 93, 85]
players = ['P1','P2','P3','P4','P5']

s = pd.Series(scores, index=players)

# Rank players
rank = s.rank(ascending=False)
print("Ranking:\n", rank)

# Top 3 players
top3 = s.sort_values(ascending=False).head(3)
print("\nTop 3:\n", top3)

# Lowest score player
lowest = s.idxmin()
print("\nLowest scorer:", lowest)
```

```
import pandas as pd

names = ['anuj', 'rahul', 'sneha', 'kiran', 'amit']
s = pd.Series(names)

# Convert to uppercase
print("Uppercase:\n", s.str.upper())

# Names containing 'a'
print("\nNames with 'a':\n", s[s.str.contains('a')])

# Replace name
s = s.replace('anuj', 'Anuj Kumar')
print("\nAfter Replace:\n", s)
```

```
import pandas as pd

data = [5, None, 15, 25, None, 35, 45]
s = pd.Series(data)

# Fill missing with median
s = s.fillna(s.median())

# Filter > 20
s = s[s > 20]

# Sort descending
s = s.sort_values(ascending=False)

# Reset index
s = s.reset_index(drop=True)

print(s)
```

```
import pandas as pd

sales = [200, 450, 300, 150, 500, 700, 100]
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

s = pd.Series(sales, index=days)

# Best and worst day
print("Best Day:", s.idxmax())
print("Worst Day:", s.idxmin())

# Total and average sales
print("\nTotal:", s.sum())
print("Average:", s.mean())

# Apply 20% bonus where sales > 400
bonus = s.copy()
bonus[bonus > 400] = bonus[bonus > 400] * 1.20
print("\nAfter Bonus:\n", bonus)

# Sort descending
print("\nSorted Sales:\n", s.sort_values(ascending=False))
```
